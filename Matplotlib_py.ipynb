```python
#setting font properties for title,label by using keyword fontdict
import numpy as np
import matplotlib.pyplot as plt
x=np.array([80,85,90,95,100,105,110,115,120,125])
y=np.array([240,250,260,270,280,290,300,310,320,330])
font1={'family':'serif','color':'blue','size':20}
font2={'family':'serif','color':'darkred','size':15}
plt.title('sports watch data',fontdict=font1)
plt.xlabel("average pulse",fontdict=font2)
plt.ylabel("Calorie burnage",fontdict=font2)
plt.plot(x,y)
plt.show()
```

```python
#we can add an title to the entire figure said as super title that is suptitle() to add title for the entire figure
import numpy as np
import matplotlib.pyplot as plt
x=np.array([0,1,2,3])
y=np.array([3,8,1,10])
plt.subplot(2,3,1)
plt.plot(x,y)
plt.title('sales')

x=np.array([0,1,2,3])
y=np.array([10,20,30,40])
plt.subplot(2,3,2)
plt.plot(x,y)
plt.title("income")
plt.suptitle("My shop", ha="center")
plt.show()
```

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(1, 2)

axs[0].plot([3, 8, 1, 10])
axs[0].set_title("sales")

axs[1].plot([10, 20, 30, 40])
axs[1].set_title("income")

# Centered super title
fig.suptitle("My shop", x=0.5, ha="center")

# Leave space for the suptitle
plt.tight_layout(rect=[0, 0, 1, 0.95])

plt.show()
```

```python
#we can even set a color for each dot by using an array of colors
import matplotlib.pyplot as plt
import numpy as np
x=np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y=np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors=np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"])
plt.scatter(x,y,c=colors)
plt.show()
```

matplotlib module has number of available colormap in which each colors has a specific values from 0 to 100 we can specify the colormap using the keyword cmap='viridis'

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y=np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.arange(len(x)) * 10  # Generate 13 color values for 13 data points
plt.scatter(x,y,c=colors,cmap='viridis')
plt.colorbar()
plt.show()
```

```python
# we can change the size of the dot with s argument
import matplotlib.pyplot as plt
import numpy as np
x=np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y=np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes=np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])
plt.scatter(x,y,s=sizes)
plt.show()
```

```python
#we can adjust the transparency of the dots with alpha argument
import matplotlib.pyplot as plt
import numpy as np
x=np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y=np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes=np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])
plt.scatter(x,y,s=sizes,alpha=0.5)
plt.show()
```

combine color size and alpha

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.random.randint(100,size=(100))
y=np.random.randint(100,size=(100))
colors=np.random.randint(100,size=(100))
sizes=10*np.random.randint(100,size=(100))
plt.scatter(x,y,c=colors,s=sizes,alpha=0.5,cmap='nipy_spectral')
plt.colorbar()
plt.show()
```

MATPLOTLIB PIE CHART

```python
import matplotlib.pyplot as plt
import numpy as np
y=np.array([35,25,25,15])
plt.pie(y)
plt.show()
```

Labels

```python
# add label to the pie chart with the label parameter
import matplotlib.pyplot as plt
import numpy as np
y=np.array([35,25,25,15])
mylabels=["apples","bannanas","cherries","dates"]
plt.pie(y,labels=mylabels)
plt.show()
```

startangle

```python
# a startangle parameter is defined with angles in degrees, default angle is 0
#eg- start the first wedge at 90 degrees
import matplotlib.pyplot as plt
import numpy as np
y=np.array([35,25,25,15])
mylabels=["apples","bannanas","cherries","dates"]
plt.pie(y,labels=mylabels,startangle=90)
plt.show()
```

Explode

```python
# if we want any of the wedges to stand out we use explode keyword
#eg-pull the "apples" wedge 0.2 from the pie:
import matplotlib.pyplot as plt
import numpy as np
y=np.array([35,25,25,15])
mylabels=["apples","bannanas","cherries","dates"]
myexplode=[0.2,0,0,0]
plt.pie(y,labels=mylabels,explode=myexplode)
plt.show()
```

Shadow

```python
# add a shadow to the pie chart by setting the shadow parameter to true
import matplotlib.pyplot as plt
import numpy as np
y=np.array([35,25,25,15])
mylabels=["apples","bannanas","cherries","dates"]
myexplode=[0.2,0,0,0]
plt.pie(y,labels=mylabels,explode=myexplode,shadow=True)
plt.show()
```

```python
# we can set the colors for each wedge using color parameter
#eg- speciy a color for each wedge
import matplotlib.pyplot as plt
import numpy as np
y=np.array([35,25,25,15])
mylabels=["apples","bannanas","cherries","dates"]
mycolors=["black","hotpink","b","#4CAF50"]
plt.pie(y,labels=mylabels,colors=mycolors)
plt.show()
```

Legend-To add a list of explanation to each wedge use legend() function

```python
import matplotlib.pyplot as plt
import numpy as np
y=np.array([35,25,25,15])
mylabels=["apples","bannanas","cherries","dates"]
plt.pie(y,labels=mylabels)
plt.legend()
plt.show()
```

Legend with header- to add a header to the legend function add a title parameter to the legend function


```python
import matplotlib.pyplot as plt
import numpy as np
y=np.array([35,25,25,15])
mylabels=["apples","bannanas","cherries","dates"]
plt.pie(y,labels=mylabels)
plt.legend(title="four fruits:")
plt.show()
```
